### 만약 mql 등록에 성공하면 seller_id가 부여되므로 이를 기준으로 seller 등록여부 열을 추가함

```python
mf['contracted'] = mf['seller_id'].apply(lambda x: 'No' if pd.isnull(x) else 'Yes')
mf['contracted'].value_counts()
```

```python
# 기존 테이블을 업데이트하려면, 수정된 DataFrame을 다시 SQL 테이블에 반영
mf.to_sql('marketing_funnel', conn, if_exists='replace', index=False)
pd.read_sql('SELECT contracted FROM marketing_funnel', conn)
```
