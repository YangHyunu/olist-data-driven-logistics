```python
from pyspark.sql import SparkSession
# Spark 세션 생성
spark = SparkSession.builder.appName("olist").getOrCreate()

df = spark.read.csv('mart_reviews.csv', header=True, inferSchema=True)
df.show()
```

```python
df.createOrReplaceTempView("df_view")
```

```python
query = """
    SELECT sentimental, AVG(review_score) as avg_review_score, COUNT(DISTINCT df_view.order_id) as cnt
    FROM df_view
    WHERE review_comment_message IS NOT NULL
    GROUP BY sentimental
    ORDER BY sentimental
"""

spark.sql(query).show()
```

- 리뷰 코멘트가 달린 후 응답까지 평균적으로 3일정도 걸림

```python
query = """
WITH CTE AS (
    SELECT review_answer_timestamp,review_creation_date, DATEDIFF(review_answer_timestamp,review_creation_date)  AS time_diff
FROM df_view
)
    SELECT AVG(time_diff) as avg_time_diff
    FROM CTE
"""
spark.sql(query).show()
```

- 리뷰 코멘트가 달린 후 3일 이내 응답한 경우, 3일 이후에 응답한 경우를 그룹으로 나눠 리뷰 점수를 테스트
- 3일 이내 응답한 경우가 3일 이후 응답한 경우보다 높은 점수를 받고, 수도 많으나 큰 차이를 보이지 않음

```python
query = """
    WITH CTE AS (
        SELECT *, CASE WHEN DATEDIFF(review_answer_timestamp,review_creation_date) > 3 THEN 'delay' ELSE 'no_delay' END as delay_flag
        FROM df_view
    )
    SELECT delay_flag,AVG(review_score) as avg_review_score, COUNT(DISTINCT order_id) as cnt 
    FROM CTE
    GROUP BY delay_flag
"""
spark.sql(query).show()
```

- 감성분석 결과 Positive의 경우 5점에 가까운 점수를 받았으나 Neutral , Negative의 경우 3점 혹은 3점 이하의 점수를 받았다.
    - 다만 무엇때문에 감성분석이 이렇게 나왔는지 분석해봐야 한다.

1. TF-IDF 벡터라이저를 통해 얻은 토큰의 스코어를 바탕으로 각 감정의 review_comment_message에 들어있는 토큰의 영향을 알아보자?


```python
query = """
    SELECT sentimental, AVG(review_score) as avg_review_score,COUNT(DISTINCT order_id) as cnt
    FROM df_view
    WHERE review_comment_message REGEXP "antes"
    GROUP BY sentimental
    ORDER BY sentimental
"""
spark.sql(query).show()
```

- 리뷰에 "antes" : 이전에(빠른)이라는 단어가 들어갔을때, 리뷰 점수가 기존에 비해 훨씬 높은 점수를 받았다. 특히 Neutral 에서 약 1.2점이상 높은 점수를 받았다.
- 그럼 "antes"가 들어간 리뷰를 받은 주문은 얼마나 걸렸을까?

```python
query = """
    SELECT CASE WHEN  review_comment_message REGEXP "antes" THEN "Yes" ELSE "No" END flag,
    AVG(review_score) as avg_review_score, COUNT(DISTINCT order_id) as cnt
    FROM df_view
    GROUP BY flag
"""
spark.sql(query).show()
```

```python
query = """
    SELECT CASE WHEN  review_comment_message REGEXP "antes" THEN "Yes" ELSE "No" END flag,
    AVG(days_to_delivery) as avg_days_to_delivery
    FROM df_view
    GROUP BY flag
"""
spark.sql(query).show()
```

- 배송기간이 약 1.2일 정도 빨라진다면 리뷰 점수가 약 0.9점 즉, 약 16%정도 증가하는 것 을 볼 수 있다.
    - 다만 리뷰 점수가 높음에 따라 평균적인 매출이 증가한다고 할 수 는 없다. 이는 아직 데이터가 부족하기 때문 
    - 리뷰 점수가 높아지면 매출이 증가할 것이라는 가설을 세우고, 이를 검증하기 위해 추가적인 데이터 수집이 필요하다.

```python
query = """
    SELECT ROUND(review_score) as avg_review_score, ROUND(AVG(price*quantity),2) as avg_sales ,COUNT(DISTINCT order_id) as cnt
    FROM df_view
    GROUP BY avg_review_score
    ORDER BY avg_sales DESC
"""
spark.sql(query).show()
```

```python
qdf = overall.select('order_id','customer_id','order_purchase_timestamp','price','quantity','product_category_name').toPandas()
qdf.head()
```

```python
qdf['payment_value'] = qdf['price'] * qdf['quantity']
qdf['order_purchase_timestamp'] = pd.to_datetime(qdf['order_purchase_timestamp'])
```

```python
rfm = qdf.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (qdf['order_purchase_timestamp'].max() - x.max()).days,
    'order_id': 'count',
    'payment_value': 'sum'
}).rename(columns={'order_purchase_timestamp': 'Recency', 'order_id': 'Frequency', 'payment_value': 'Monetary'})
```

```python
rfm.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 데이터 준비
rfm_data = rfm[['Recency', 'Frequency', 'Monetary']]
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_data)

# K-Means 클러스터링 적용
kmeans = KMeans(n_clusters=3, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# 결과 확인
rfm.groupby('Cluster').mean()
```

```python
rfm.head()
```

```python
rfm_orders =rfm.merge(df,on='customer_id',how='inner')
rfm_orders.head()
```

```python
category_count = rfm_orders.groupby('Cluster')['product_category_name'].value_counts().unstack().fillna(0)
category_count.head()
```
